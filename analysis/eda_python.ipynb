```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Setting the visual style for charts
sns.set_theme(style="whitegrid")

# Loading survey data and schema
df = pd.read_csv('survey_results_public.csv', low_memory=False)
schema_df = pd.read_csv('survey_results_schema.csv')
pd.set_option('display.max_columns', None)

print(df.shape)
print(schema_df.shape)
display(df.head())
display(schema_df.head())
```

```python
# --- DATA CLEANING ---


# Identifying missing values across columns
# Sorting helps highlight columns with the most missing data
missing_values = df.isna().sum().sort_values(ascending=False)
print(missing_values[missing_values > 0].head(20))
print("\nTotal columns with missing values:", (missing_values > 0).sum())

# Counting total records in 'ResponseId'
print(f"\nTotal records: {df['ResponseId'].count()}")

# Dropping duplicates in 'ResponseId' if there are any
df = df.drop_duplicates(subset='ResponseId')

# Verifying the count after dropping duplicates
print(f"Total records after removing duplicates: {df['ResponseId'].count()}\n")
```

```python
# --- ANALYSIS ---

#1. Total number of respondents
print(f"Total number of respondents: {df['ResponseId'].count()}\n")
```

```python
#2. Analyzing distribution of respondents by education level

df_education = df.dropna(subset=['EdLevel'])
df_education = (
    df_education
    .groupby('EdLevel')
    .size()
    .reset_index(name='respondent_count')  
    .sort_values(by='respondent_count', ascending=False)  
)

display(df_education)
```

```python
#3. Analyzing distribution of years of professional experience
# Missing values are removed to avoid skewing the distribution

work_exp = df['WorkExp'].dropna()
work_exp = work_exp[work_exp <= 50] # remove unrealistic values

experience = pd.Series({
    'Mean': round(work_exp.mean(), 1),
    'Median': round(work_exp.median(), 1),
    'Mode': round(work_exp.mode().iloc[0], 1)
})

print("Work Experience Statistics:\n")
print(experience)
print("\n")
```

```python
# --- Visualization: Distribution of Respondents Work Experience

plt.figure(figsize=(10, 5))

sns.histplot(work_exp, bins=30, kde=True)

plt.title('Distribution of Work Experience', fontsize=15, fontweight='bold')
plt.xlabel('Years of Work Experience', fontweight='bold')
plt.ylabel('Number of Respondents', fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
#4. Analyzing work setup: remote, hybrid, and flexible arrangements

# Calculating exclusively remote workers
remote = (df['RemoteWork'] == 'Remote').sum()

# Calculating hybrid or flexible workers, excluding fully remote workers
hybrid_or_flexible = df['RemoteWork'].isin([
    'Hybrid (some in-person, leans heavy to flexibility)',
    'Hybrid (some remote, leans heavy to in-person)',
    'Your choice (very flexible, you can come in when you want or just as needed)'
]).sum()

# Calculating all respondents with any remote/flexible option
remote_or_flexible = remote + hybrid_or_flexible

print(f"\nNumber of respondents working exclusively remote: {remote}")
print(f"Number of respondents working hybrid or flexible: {hybrid_or_flexible}")
print(f"Number of respondents with any remote/flexible option: {remote_or_flexible}\n")
```

```python
#5. Calculating how many respondents have experience with Python

python_users = df['LanguageHaveWorkedWith'].str.contains('Python', case=False, na=False).sum()
respondent_total = df['ResponseId'].nunique()
python_percentage = (python_users / respondent_total) * 100

print(f"Number of respondents who code in Python: {python_users}")
print(f"Percentage of respondents who code in Python: {round(python_percentage, 1)} %")
```

```python
#6. Geographical analysis of Python developers' compensation

# Filtering Python developers for salary analysis
# Removing missing values
# Filtering out extreme outliers to improve reliability
python_salary_df = df[df['LanguageHaveWorkedWith'].str.contains('Python', case=False, na=False)].copy()
python_salary_df = python_salary_df.dropna(subset=['ConvertedCompYearly', 'Country'])
python_salary_df = python_salary_df[
    python_salary_df['ConvertedCompYearly'].between(1000, 500000)
]

# Calculating salary statistics by country (mean, median, respondent count)
python_country_salary = (
    python_salary_df
    .groupby('Country')['ConvertedCompYearly']
    .agg(
        mean='mean',
        median='median',
        respondent_count='count'
    )
    .round()
    .reset_index()
)

# Excluding countries with fewer than 20 Python developer respondents
python_country_salary = python_country_salary[python_country_salary['respondent_count'] >= 20]

python_country_salary = python_country_salary.sort_values(by='median', ascending=False)

display(python_country_salary)
```

```python
# --- Visualization: Top 10 Countries by Median Salary (Python) ---

# Median is used instead of mean to reduce the impact of salary outliers

top_10_countries = python_country_salary.sort_values(by='median', ascending=False).head(10)

plt.figure(figsize=(12, 6))

sns.barplot(
    x='median',
    y='Country',
    data=top_10_countries,
    hue='Country',
    palette='plasma',
    legend=False
)

plt.title('Top 10 Countries by Median Salary for Python Developers', fontsize=15, fontweight='bold')
plt.xlabel('Median Yearly Compensation (USD)', fontweight='bold')
plt.ylabel('Country', fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
#7. Education analysis of the highest-paid Python specialists

# Filtering only Python developers
education_python = df[df['LanguageHaveWorkedWith'].str.contains('Python', case=False, na=False)].copy()

# Removing rows with missing salary or education data
# Filtering out extreme outliers 
education_python = education_python.dropna(subset=['ConvertedCompYearly', 'EdLevel'])
education_python = education_python[
    education_python['ConvertedCompYearly'].between(1000, 500000)
]

# Sorting by salary (descending)
education_python = education_python.sort_values(by='ConvertedCompYearly', ascending=False)
education_python = education_python.reset_index(drop=True)
education_python.index += 1

# Selecting top 100 highest-paid Python developers for further analysis
top_100_python_earners = education_python.head(100)

display(top_100_python_earners[['ResponseId', 'ConvertedCompYearly', 'EdLevel']].head(10))
```

```python
# --- Visualization: Education Levels of Top Earners ---

# Counting education levels among the top 100 highest-paid Python developers
# Aggregating counts by education level
education_top_earners = (
    top_100_python_earners
    .groupby('EdLevel')
    .size()
    .reset_index(name='respondent_count')
    .sort_values(by='respondent_count', ascending=False)
)

plt.figure(figsize=(14, 6))

sns.barplot(
    data=education_top_earners,
    x='respondent_count',
    y='EdLevel',
    hue='EdLevel',
    palette='magma',
    legend=False
)

plt.title('Education Levels of the Top 100 Highest-Paid Python Developers',
          fontsize=14, fontweight='bold')
plt.xlabel('Number of Respondents', fontsize=12, fontweight='bold')
plt.ylabel('Education Level', fontsize=12, fontweight='bold')

plt.tight_layout()
plt.show()
```
